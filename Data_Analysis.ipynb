**Step3 (_Combine_):** We can also combine the results of the analysis into a new data structure(s).

# 3. Matplotlib

## 3.1 pyplot

matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure e.g. creates a figure, plots some lines in a plotting area, decorates the plot with labels, etc. In matplotlib.pyplot various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current subplot.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,5])
plt.ylabel('some numbers')
```

plot() is a versatile command, and will take an arbitary number of arguments. For example, to plot x versus y, you can issue the command:

```python
plt.plot([1,2,3,4], [1,4,9,16])
```

Using the pyplot interface, you build a graph by calling a sequence of functions and all of them are applied to the current subplot, like so:

```python
plt.plot([1,2,3,4], [10,20, 30, 40], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.title('Title of the plot')
plt.xlabel('This is the x-label')
plt.ylabel('This is the y-label')
```

When working with just one subplot in the figure, generally it is OK to work with the pyplon interface, however, when doing more complicated plots, or working with larger scripts, you will want to explicitly pass around the Subplot (Axes) and/or Figure object to operate upon.

```python
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t1, f(t1), 'bo')

plt.subplot(2, 1, 2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.ylabel('Y label') # Which subplot is modifying this funcion
```

The advantage of keeping which subplot we're working with very clear in our code is obvious when we start to have multiple axes in one figure. Because explicit is better than implicit I recommend working with the Object Oriented interface.

## 3.2 Object Oriented Interface

What you do is to create objects and then call methods for every object to make changes specifically to that object. The _plt.subplots_ function creates a figure and a Subplot (Axes) object at the same time, then we can create a plot by calling methods on those objects.

```python
x = np.linspace(-4, 4)
y1, y2 = x**2, x**3
```

```python
# Object Oriented Interface
fig, ax = plt.subplots()
ax.plot(x, y1, 'red')
ax.plot(x, y2, 'blue')
ax.set_title('Quadratic and Cubic Powers')
ax.set_xlabel('x values')
ax.set_ylabel('y values')
```

Interestingly, just about any methods of an Axes object exist as a function in the pyplot module (and vice-versa). For example, when calling plt.xlabel('x label'), pyplot calls ax.set_xlabel('x label') on whichever Axes is 'current'.

```python
# pyplot
plt.plot(x, y1, 'red', x, y2, 'blue')
plt.title('Quadratic and Cubic Powers')
plt.xlabel('x values')
plt.ylabel('y labels')
```

The usefulness of the OOI is clearer when working with multiple subplots in one figure. A figure can have more than one _Axes_ on it. If you want your axes to be on a regular grid system, then it's easiest to use _plt.subplots(...)_ to create a figure and add the axes automatically.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

`plt.subplots(...)` created a new figure and added 4 subplots to it. The axes object that was returned is a 2D numpy object array. Each item in the array is one of the subplots. When we want to work with one of these axes, we can index the `axes` array and use the item's methods.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].set_title('Upper Left')
axes[0,1].set_title('Upper Right')
axes[1,0].set_title('Lower Left')
axes[1,1].set_title('Lower Right')

# To iterate over all items in a multidimensional numpy array, use the `flat` attribute
for ax in axes.flat:
    # Remove all xticks and yticks
    ax.set(xticks=[], yticks=[])

fig.tight_layout()
```

### 3.2.1 Example: Creating a figure showing the shape of the power functions

```python
x = np.linspace(start=-5, stop=5, num=150)
```

```python
# All functions in one axes
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(x, x, label='Linear')
ax.plot(x, x**2, label='Quadratic')
ax.plot(x, x**3, label='Cubic')
ax.plot(x, x**4, label='4th Power')
ax.legend()
```

```python
# 4 Axes/Subplots: One function in one axes
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7,4.5))
axes[0,0].set_title('Linear')
axes[0,0].plot(x, x)
axes[0,1].set_title('Quadratic')
axes[0,1].plot(x, x**2)
axes[1,0].set_title('Cubic')
axes[1,0].plot(x, x**3)
axes[1,1].set_title('4th Power')
axes[1,1].plot(x, x**4)
fig.tight_layout()
```

What if we want to take a look at the first 10 powers of x?

```python
# A more elegant approach
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(14,4.5))
for i, ax in enumerate(axes.flatten()):
    ax.set_title('x to the {:d}'.format(i+1))
    ax.plot(x, x**(i+1))

fig.tight_layout()
```

# 4. Common customizations: colors, labels, ticks, tickmarks, limits and annotations

Matplotlib allows the user to customize almosh every element of a plot. However, there are some elements that we change more often, those are the ones that we review here.

```python
# Generating data
x = np.linspace(-np.pi, np.pi, 200)
sine, cosine = np.sin(x), np.cos(x)
```

In this section we will go from this simple plot:

```python
fig, ax = plt.subplots()
ax.plot(x, sine, x, cosine)
```

To this plot:

```python
fig, ax = plt.subplots()

# 4.1 Colors
# Can be given as string, hex value, gray scale, or RGB[A] tuples
ax.plot(x, sine, color='red', label='Sine')
ax.plot(x, cosine, color='#165181', label='Cosine')

# 4.2 Setting axis limits
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-1.2, 1.2)

# 4.3 Setting ticks and tick labels
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1,0,1])

ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'], size=17)
ax.set_yticklabels(['-1', '0', '+1'], size=17)

# 4.4 Legend
ax.legend(loc='upper left')

# 4.5 Annotations
ax.text(-0.25,0,'(0,0)') # x coord, y coord
ax.text(np.pi-0.2,0.05, r'$(\pi,0)$', size=15)

ax.annotate('Origin',
            xy=(0,0), # where the arrow points to
            xytext=(1, -0.7), # location of text
            arrowprops=dict(facecolor='blue'))

# 4.6 Grids, horizontal and vertical lines
ax.axhline(0, color='black', alpha=0.9) # horizontal line
ax.axvline(0, color='black', alpha=0.9) # vertical line

ax.grid()
```

```python
def identify_cat_above30(series):
    counts = series.value_counts()
    return list(counts[counts > 30].index)
```

### 5.2.7 Relationship of SalePrice with categorical variables

The standard plot to examine the relationship between a numerical and a categorical variable is the boxplot. #### **Confidence Intervals**

**Confidence interval for the mean of the final grade**

We can calculate confidence intervals for the means and for proportions,

```python
sample_size = student.shape[0]
print(sample_size)
```

Because we have a sample size that is much greater than 30, we can use the Central Limit Theorem to calculate confidence intervals. According to the theorem we can calculate a confidence interval for the mean using the normal distribution.

To get the confidence interval for the mean we need three numbers:
1. Sample mean
2. Standard error
3. Confidence level

Formula for the standard error:
$$SE = \frac{s}{\sqrt{n}}$$

```python
sample_mean_grade = student['G3'].mean()
sample_mean_grade
```

```python
std_error_grades = student['G3'].std()/sqrt(sample_size)
```

```python
stats.norm.interval(0.95, loc=sample_mean_grade, scale=std_error_grades)
```

Now let's calculate a confidence interval for the proportion of students with High Alcohol Consumption Level. Again we need 3 numbers:
1. Sample proportion
2. Standard error
3. Confidence level
For proportions the standard error is given by:
$$SE = \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

```python
student['acl'].value_counts(normalize=True)
```

```python
high_prop = student['acl'].value_counts(normalize=True)['High']
std_error_prop = sqrt(high_prop*(1-high_prop)/sample_size)
```

```python
stats.norm.interval(0.98, loc=high_prop, scale=std_error_prop)
```

0.25 (or 25.0%) seems like a good guess for the proportion of students with High Alcohol Consumption Levels in the population.

###### **Null Hypothesis Testing framework**

**1. Setting up two competing hypotheses**  
**2. Set in advance some level of significance, called alpha**  
**3. Calculate a test statistic and the p-value (or find rejection region)**  
**4. Make a test decision about the null hypothesis**  
**5. State an overall conclusion**


##### **Are the population variances equal in the two groups of students (Low vs. High alcohol consumption)?**

```python
student.groupby('acl')['G3'].var()
```

```python
grades_low_acl = student['G3'][student['acl']=='Low']
grades_high_acl = student['G3'][student['acl']=='High']
stats.bartlett(grades_low_acl, grades_high_acl)
```

According to the test we cannot reject the Null hypothesis of equal variances, so we will assume that the two groups are samples from a population with the same variances. This info will be useful in our next test.

##### **Does alcohol consumption affect academic performance?**

```python
fig, axes = plt.subplots(1, 2, figsize=(14,4))
sns.boxplot(x='acl', y='G3', data=student, ax=axes[0])
sns.pointplot(x='acl', y='G3', data=student, ax=axes[1])
```

The visualization suggests there is a difference between the means of the final grade of the two groups. Now we will perform a formal statistical test to confirm the hypothesis that students with High alcohol consumption level perform worse than the students with Low alcohol consumption level.

> **Null Hypothesis: for both groups (High and Low ACL) the population means of the final grade are equal.**  
> **Alternative Hypothesis: The population means of the final grades are different.**  

A common test to apply for these cases is the two-sample t-test, which is used to determine if two population means are equal.  
All statistical tests have auumptions that must be checked for their conclusions to be valid. For this test the assumptions are:  
1. **Independent samples**: we will assume that the method for collecting the data assured that tho answers given by the students are independent.
2. **Large enough sample size or observations come from a normally-distributed population**: this assumption is required if we are working with small samples (less than 30). Since in the smaller group we are working with 166 observations we can say that we have a "large enough" sample.
3. **Variances are equal.**  
In addition, this test has two versions: one assuming equal variances and the other assuming unequal variances.  
According to the test we can reject the Null hypothesis of equal variances, so we will assume that the two samples come from a population with the same population variances. Time to perform our t-test:

```python
stats.ttest_ind(grades_low_acl, grades_high_acl,equal_var=True)
```

Since we get such a low p-value we can reject the Null hypothesis of equal means for the two groups at a level of significance of 5%.

> **Conclusion: there is a statistically significant difference between the grades in the two analyzed groups. Since the mean for the group with high alcohol consumption is less than the mean of the other group, the results suggest that alcohol consumption has a negative impact on students' academic performance.**
